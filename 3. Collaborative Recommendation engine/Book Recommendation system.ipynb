## Recommendation system:
A piece of code which is intelligent enough to understand the user interset and suggest the products accordingly. 


### Types: 
#### 1. Popularity based:
Displaying top N products based on the reviews or votes obtained.
#### 2. Content based:
Suggests the similar kind of content to the user based on the user search.
#### 3. Collaborative filter:
It attempts to find users that have similar preferences and opinions as the input and then recommends items that they have liked to the input.
#### 4. Hybrid recommendation system.
Hybrid recommendation system combines both content based and collaborative based recommendations

### Popularity based recommendation system

Criteria followed:
Display the top 250 rated books, based on the user ratings. Consider the books with minimum votes greater than 250

```python
books_ratings = ratings.merge(books, on='ISBN')
books_ratings.shape
```

```python
books_ratings.head()
```

```python
num_ratings = books_ratings.groupby('Book-Title').agg('count')['Book-Rating'].reset_index()
num_ratings.rename(columns={'Book-Rating':'Count-rating'},inplace=True)
```

```python
avg_ratings = books_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={"Book-Rating":"Avg-rating"},inplace=True)
```

```python
books_top_ratings = num_ratings.merge(avg_ratings,on='Book-Title')
books_top_ratings.head()
```

```python
top_50_books = books_top_ratings[(books_top_ratings['Count-rating']>=250)].head(50)
top_50_books = top_50_books.sort_values(by='Avg-rating',ascending=False)
top_50_books.shape
```

```python
top_50_books_details = top_50_books.merge(books,on='Book-Title')
top_50_books_details = top_50_books_details.drop_duplicates('Book-Title')
top_50_books_details.shape
```

### Collaborative recommender system

```python
# Merging the data corresponding the movies and their ratings
book_ratings = ratings.merge(books,on='ISBN')
book_ratings.shape
```

In the collaborative recommendation model, we are going to consider the books that are having at least 50 reviews from the users and active users who gave ratings to atleast 300 books. 

```python
# Filtering the users who gave atleast 300 reviews 
x =book_ratings.groupby('User-ID')['Book-Rating'].count()>=300
filtered_users = x[x].index
filtered_book_ratings = book_ratings[book_ratings['User-ID'].isin(filtered_users)]
filtered_book_ratings.shape
filtered_book_ratings.shape
```

```python
# Filtering the books that are having minimum 50 ratings 
y =filtered_book_ratings.groupby('Book-Title')['Book-Rating'].count()>=50
filtered_books = y[y].index
filtered_book_ratings = filtered_book_ratings[filtered_book_ratings['Book-Title'].isin(filtered_books)]
filtered_book_ratings.shape
```

```python
# Creating a pivot table such that the index will be the Book title and columns are the Users
pivoted_data = filtered_book_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pivoted_data
```

```python
# Replace all the NaN with 0
pivoted_data.fillna(0,inplace=True)
pivoted_data.isna().sum()
```

```python
# Calculating the similarity  score for each row
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pivoted_data)
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pivoted_data.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[:5]
    for i in similar_books:
        similar_book_name = pivoted_data.index[i[0]]
        print(similar_book_name) 
    
```

```python
recommend('2nd Chance')
```
